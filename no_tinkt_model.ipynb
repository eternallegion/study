```python
import torch
import torch.nn as nn
import torch.optim as optim
import math
import torchaudio.transforms as ts
import torchaudio.functional as f

class model(nn.Module):
    def __init__(self):
        super(model,self).__init__()
        self.layer=nn.Sequential(
            #nn.Linear(input,32),
            nn.Linear(32,64),
            nn.ReLU(),
            nn.Dropout(),
            nn.Linear(64,128),
            nn.ReLU(),
            nn.Dropout(),
            nn.Linear(128,256),
            nn.ReLU(),
            nn.Linear(256,512),
            nn.ReLU(),
            nn.Linear(512,128),
            nn.ReLU(),
            nn.Dropout(),
            nn.Linear(128,10),
            nn.ReLU(),
            nn.Dropout(),
        )

    
    def forward(self,x):
        x=self.layer(x)
        return x

class EarlyStopping:
    def __init__(self, patience=7, delta=0.01):
        self.patience = patience
        self.delta = delta
        self.best_loss = None
        self.early_stop = False
        self.counter = 0

    def __call__(self, val_loss):
        if self.best_loss is None:
            self.best_loss = val_loss
        elif val_loss > self.best_loss - self.delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_loss = val_loss
            self.counter = 0

device=torch.device('cuda'if torch.cuda.is_available()else'cpu')

early_stopping = EarlyStopping(patience=5, delta=0.01)


model=model().to(device)
opt=optim.Adam(model.parameters(),lr=0.001)
loss_f=nn.CrossEntropyLoss()
x = torch.randn(64, 32).to(device)

# 더미 목표 라벨 (10개의 클래스 가정)
y = torch.randint(0, 10, (64,)).to(device)
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y = train_test_split(x, y, test_size=0.2)

train_x.to(device)
val_x.to(device)
train_y.to(device)
val_y.to(device)

# 훈련 루프
for epoch in range(1000):
    model.train()
    output = model(train_x)
    loss = loss_f(output, train_y)
    opt.zero_grad()
    loss.backward()
    opt.step()    
    print(f'train:Epoch {epoch}, Loss: {loss.item()}')
    model.eval()
    with torch.no_grad():
        val_output=model(val_x)
        val_loss=loss_f(val_output,val_y)


    early_stopping(val_loss)
    if early_stopping.early_stop:
        print('stop')
        break
    print(f'Epoch {epoch}, Loss: {loss.item()}, Val Loss: {val_loss.item()}')    
```
