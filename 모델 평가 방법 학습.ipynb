# 모델 평가 방법 학습

1. Accuracy (정확도)
   
의미:
전체 샘플 중에서 모델이 올바르게 예측한 샘플의 비율입니다.



해석:
예를 들어, Accuracy가 95%이면, 모델이 전체 데이터의 95%를 올바르게 분류했다는 의미입니다.

활용 방법:

전체 데이터 중 모델이 올바르게 분류한 비율을 나타내므로, 모델의 전반적인 성능을 빠르게 파악할 수 있습니다.

주의점:

클래스 불균형 문제가 있는 경우, 높은 Accuracy가 반드시 좋은 성능을 의미하지 않을 수 있습니다.

실무 활용 예:

초기 모델 평가 단계에서 모델이 전체적으로 얼마나 잘 작동하는지 확인.

데이터가 균형 잡힌 경우 전반적인 성능 지표로 활용.

## 2. Confusion Matrix (혼동 행렬)
활용 방법:

각 클래스별로 실제 라벨과 예측 라벨의 분포를 보여줍니다.

어떤 클래스에서 오분류가 많이 발생하는지, 특정 클래스끼리 혼동되는지를 시각적으로 파악할 수 있습니다.

실무 활용 예:

특정 클래스에서 지속적으로 오분류가 발생한다면, 그 클래스의 데이터가 부족하거나 특징 추출에 문제가 있을 수 있으므로, 데이터 보강, 클래스 가중치 조정, 혹은 모델 아키텍처 개선을 고려할 수 있습니다.

모델 디버깅 단계에서 어떤 클래스 간의 혼동이 심한지 분석하여 추가적인 전처리나 데이터 증강 전략을 세울 때 사용.

3. Classification Report (분류 리포트)
의미:
각 클래스별로 정밀도(Precision), 재현율(Recall), F1-score, 그리고 지원(지원 개수, Support) 등의 평가 지표를 요약한 보고서입니다.



해석:
이 리포트를 통해 각 클래스별로 모델의 성능이 어떤지, 예를 들어 어떤 클래스에서는 정밀도는 높으나 재현율이 낮은지, 또는 전반적으로 균형 잡힌 성능을 보이는지 등을 알 수 있습니다.

### Precision (정밀도)
활용 방법:

모델이 특정 클래스로 예측한 샘플 중 실제로 그 클래스에 속하는 비율을 나타냅니다.

False Positive가 중요한 상황에서, 즉 잘못 분류되었을 때 비용이 큰 경우 (예: 스팸 메일 필터링, 질병 진단 등) 높은 Precision이 중요합니다.

실무 활용 예:

모델이 특정 클래스를 예측할 때 불필요한 오분류를 줄이기 위해 정밀도를 높이는 방향으로 조정할 수 있습니다.

분류 임계치를 조정하거나, 클래스별 손실 가중치를 조정할 때 활용.


### Recall (재현율)
활용 방법:

실제 해당 클래스에 속하는 샘플 중 모델이 올바르게 예측한 비율을 나타냅니다.

False Negative가 중요한 상황에서 (예: 질병 검진, 사기 탐지 등) 높은 Recall이 중요합니다.

실무 활용 예:

모델이 실제 양성 사례를 많이 놓치고 있다면(낮은 Recall), 모델의 민감도를 높이도록 조정할 수 있습니다.

데이터 증강이나 모델의 구조를 조정하여 재현율을 개선할 수 있습니다.

### F1-score
활용 방법:

Precision과 Recall의 조화 평균으로, 두 지표 간의 균형을 평가합니다.

한쪽으로 치우치지 않는 균형 잡힌 성능 평가가 필요한 경우 유용합니다.

실무 활용 예:

Precision과 Recall 둘 다 중요한 상황에서, F1-score를 최적화하는 방향으로 모델을 개선할 수 있습니다.

모델 비교 시, 단순 Accuracy보다 더 세밀한 성능 평가 기준으로 사용.


### Classification Report
활용 방법:

각 클래스별로 Precision, Recall, F1-score, Support(실제 샘플 수)를 종합적으로 보여줍니다.

각 클래스별 성능을 한눈에 확인할 수 있어, 어느 클래스에서 모델이 약한지 명확하게 파악할 수 있습니다.

실무 활용 예:

클래스 불균형 문제를 진단하고, 특정 클래스에 대한 성능 개선이 필요한지 결정하는 데 사용합니다.

모델 튜닝 및 개선 방향을 정할 때, 각 클래스별로 어떤 지표가 낮은지 보고 대응 전략(예: 데이터 추가, 손실 가중치 조정 등)을 세울 수 있습니다.


# 종합 활용 방안
모델 개선 방향 결정:

전체 Accuracy로 기본 성능을 확인한 후, Confusion Matrix와 Classification Report를 통해 어느 클래스에서 문제가 발생하는지 파악합니다.

Precision, Recall, F1-score를 종합적으로 고려해 특정 클래스에 대한 모델의 약점을 진단하고, 추가적인 전처리, 데이터 증강, 혹은 모델 수정(예: 클래스 가중치 조정, 임계치 조정)을 시행합니다.

하이퍼파라미터 튜닝:

학습률, 배치 크기 등의 하이퍼파라미터를 조정하면서, 이러한 평가 지표들이 어떻게 변하는지 모니터링합니다.

평가 지표의 변화에 따라 모델이 과대적합(Overfitting) 또는 과소적합(Underfitting)되는지 진단하고, 조정할 수 있습니다.

```python
import torch
import torch.nn as nn 
import torch.optim as optim
from torch.utils.data import DataLoader as loader
from torchvision import datasets, transforms
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
```

```python
class simple(nn.Module):
    def __init__(self):
        super(simple,self).__init__()
        self.fc1=nn.Linear(28*28, 128)
        self.fc2=nn.Linear(128,10)

    def forward(self, x):
        x=x.view(-1,28*28)
        x=torch.relu(self.fc1(x))
        x=self.fc2(x)
        return x
```

```python
transform=transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.1307,),(0.3081))])

test_data=datasets.MNIST(root="./data",train=False, download=True, transform=transform)
test_loader=loader(test_data, batch_size=1000,shuffle=True)

model=simple()
```

```python
def eval_matric(model, test_loader):
    model.eval()
    all_pred=[]
    all_target=[]

    with torch.no_grad():
        for data, target in  test_loader:
            out=model(data)
            pred=out.argmax(dim=1)
            all_pred.extend(pred.cpu().numpy())
            all_target.extend(target.cpu().numpy())

    all_pred=np.array(all_pred)
    all_target=np.array(all_target)

    accuracy=np.mean(all_pred==all_target)*100  #정확도
    comf_mat=confusion_matrix(all_target,all_pred)
    report=classification_report(all_target, all_pred)
    return accuracy, comf_mat, report
```

```python
acc, conf_mat, report=eval_matric(model, test_loader)
print(f"Test Accuracy: {acc:.2f}%")
print("\nConfusion Matrix:")
print(conf_mat)
print("\nClassification Report:")
print(report)
```

```python
#혼동행렬을 시각화

plt.figure(figsize=(8, 6))  #그래프 사이즈

plt.imshow(conf_mat, interpolation='nearest', cmap=plt.cm.Blues)  
#설명:
#conf_mat이라는 혼동 행렬 데이터를 이미지 형태로 보여줍니다.
#interpolation='nearest'는 데이터의 각 셀을 개별 블록으로 표현하여 값의 경계를 명확하게 보이도록 합니다.
#cmap=plt.cm.Blues는 색상 맵을 파란색 계열로 설정해, 값이 클수록 진한 파란색으로 표시되도록 합니다.

plt.title("Confusion Matrix")
plt.colorbar()
#오른쪽에 색상 막대(colorbar)를 추가하여, 색상이 어떤 수치 값을 나타내는지 시각적으로 확인할 수 있게 합니다.

tick_marks = np.arange(10)
#설명:
#np.arange(10)는 0부터 9까지의 숫자 배열을 생성합니다.
#여기서는 MNIST처럼 10개의 클래스를 가정하여, 각 클래스 번호를 tick mark(눈금)로 사용합니다.
plt.xticks(tick_marks, tick_marks)
plt.yticks(tick_marks, tick_marks)
#설명:
#x축과 y축의 눈금을 설정합니다.
#첫 번째 인자는 눈금의 위치, 두 번째 인자는 그 위치에 표시할 레이블입니다.
#여기서는 0부터 9까지의 숫자를 그대로 사용하여, 각 축에 해당 클래스 번호를 표시합니다.

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
#히트맵 버전
import seaborn as sns

sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

# 히트맵(Heatmap)이란?
정의:
행렬(또는 2차원 배열)의 각 요소를 색상의 강도나 색조로 표현하는 시각화 기법입니다.

값이 클수록 진하거나 밝은 색으로 표시되고, 값이 작을수록 연하거나 어두운 색으로 표현됩니다.

onf_mat:
시각화할 행렬 데이터(여기서는 혼동 행렬).

interpolation='nearest':
각 픽셀(행렬의 요소)을 그대로 보여주어, 값의 경계를 명확하게 합니다.
이를 통해, 각 셀의 색상이 값의 크기를 명확히 표현할 수 있습니다.

cmap=plt.cm.Blues:
'Blues'라는 색상 맵을 사용해, 값이 클수록 진한 파란색, 값이 작을수록 연한 파란색으로 표현합니다.
