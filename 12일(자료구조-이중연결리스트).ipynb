이중 연결 리스트란?
 
    -각 노드가 이전 노드와 다음 노드에 대한 포인터를 가짐.
 
    -단일 연결 리스트보다 메모리 사용량이 많지만, 양방향 탐색이 가능.

    각 노드: 데이터, prev(이전 노드), next(다음 노드).

```python
class DoubleNode:
    def __init__(self, data):
        self.data=data
        self.next=None
        self.prev=None
```

```python
class DoubleLinkedList:
    def __init__(self):
        self.head=None
        self.tail=None

    def append(self, data):
        new_node=DoubleNode(data)
        if self.head is None:
            self.head=new_node
            self.tail=new_node
        else:
            self.tail.next=new_node
            new_node.prev=self.tail
            self.tail=new_node

    def display_forward(self):
        current=self.head
        while current:
            print(current.data,end="<->")
            current=current.next
        print("None")

    def display_backward(self):
        current=self.tail
        while current:
            print(current.data,end="<->")
            current=current.prev
        print("None")
```

```python
dl=DoubleLinkedList()
dl.append(1)
dl.append(2)
dl.append(3)
dl.append(4)
dl.append(5)
print(dl.display_forward())
print(dl.display_backward())
```
