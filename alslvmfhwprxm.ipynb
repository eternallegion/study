```python
class User:
    def __init__(self, name, email, password):
        self.name=name
        self.email=email
        self.password=password
        self.cart=Cart(self)

    def __str__(self):
        return f"User: {self.name}, Email: {self.email}"

class Product:
    def __init__(self,name, price,stock):
        self.name=name
        self.price=price
        self.stock=stock

    def __str__(self):
        return f"{self.name} - ${self.price} (Stock: {self.stock})"

class Cart:
    def __init__(self, user):
        self.user=user
        self.items={}

    # """ÏÉÅÌíàÏùÑ Ïû•Î∞îÍµ¨ÎãàÏóê Ï∂îÍ∞Ä"""
    def add_to_cart(self, product, quantity):
        if product.stock>quantity:
            self.items[product]=self.items.get(product,0)+quantity
            product.stock-=quantity
            print(f"{product.name} {quantity}Í∞ú Ïû•Î∞îÍµ¨ÎãàÏóê Ï∂îÍ∞ÄÎê®.")
        else:
            print(f"{product.name}Ïùò Ïû¨Í≥†Í∞Ä Î∂ÄÏ°±Ìï©ÎãàÎã§.")

    def remove_from_cart(self,product, quantity):
        if product in self.items:
            if self.items[product]>=quantity:
                self.items[product]-=quantity
                product.stock+=quantity
                print(f"{product.name} {quantity}Í∞ú Ï†úÍ±∞Îê®.")
            else:
                product.stock+=self.items[product]
                del self.items[product]
                print(f"{product.name}Ïù¥(Í∞Ä) Ïû•Î∞îÍµ¨ÎãàÏóêÏÑú Ï†úÍ±∞Îê®.")

        else:
            print(f"{product.name}Ïù¥(Í∞Ä) Ïû•Î∞îÍµ¨ÎãàÏóê ÏóÜÏäµÎãàÎã§.")

 #"""Ïû•Î∞îÍµ¨Îãà Î™©Î°ù Ï∂úÎ†•"""
    def view_cart(self):
        print(f"\nüõí {self.user.name}Ïùò Ïû•Î∞îÍµ¨Îãà:")
        if not self.items:
            print("Ïû•Î∞îÍµ¨ÎãàÍ∞Ä ÎπÑÏñ¥ ÏûàÏäµÎãàÎã§.")
        else:
            for product,quantity in self.items.items():
                print(f"- {product.name}: {quantity}Í∞ú (${product.price} √ó {quantity} = ${product.price * quantity})")


class Order:
    def __init__(self,user,cart):
        self.user=user
        self.cart=cart
        self.total=sum(p.price * q for p, q in cart.items.items())
    def checkout(self):
        print(f"\n‚úÖ {self.user.name}ÎãòÏùò Ï£ºÎ¨∏Ïù¥ ÏôÑÎ£åÎêòÏóàÏäµÎãàÎã§. Ï¥ù Í∏àÏï°: ${self.total}")
        self.cart.items.clear()
```

```python
class Book:
    def __init__(self, title, author, isbn, stock):
        self.title=title
        self.author=author
        self.isbn=isbn
        self.stock=stock

    def __str__(self):
        return f"{self.title} - {self.author} (ISBN: {self.isbn})"


class Library:
    def __init__(self):
        self.data={}

    def add_book(self, book):
        if book.isbn in self.data:
            EXbook=self.data[book.isbn]
            EXbook.stock+=book.stock
            print(f"Updated stock for {book.title}. New stock: {EXbook.stock}")
        else:
            self.data[book.isbn]=book
            print(f"‚úÖ {book.title} Ï∂îÍ∞Ä ÏôÑÎ£å!")
            print(f"{book}")

    def finds_book(self, quary):
        if quary in self.data:
            return self.data[quary]
        for book in self.data.values():
            if book.title.lower()==quary.lower():
                return book
        print("Book not found.")
        return None
    

    def remove_book(self, ISBN):
        if ISBN in self.data:
            del self.data[ISBN]
            print(f"Removed book with ISBN: {ISBN}")
        else:
            return "cheak"

    def borrow_book(self, isbn):
        book=self.data.get(isbn)
        if book:
            if book.stock>0:
                book.stock-=1
                print(f"Borrowed book: {book.title}. Remaining stock: {book.stock}")
            else:
                print(f"{book.title} is not available for borrowing (out of stock).")
        else:
            print("Book not found.")

    def return_book(self,isbn):
        book=self.data.get(isbn)
        if book:
            book.stock+=1
            print(f"Returned book: {book.title}. New stock: {book.stock}")
        else:
            print("Book not found in library system.")

    def list_books(self):
        if not self.data:
            print("Library is empty.")
        else:
            print("Library Books:")
            for book in self.data.values():
                print(book)
```

```python
library = Library()

    # 2. Book Í∞ùÏ≤¥ ÏÉùÏÑ±
book1 = Book("Python Programming", "John Doe", "978-1234567890", 3)
book2 = Book("Data Structures", "Jane Smith", "978-0987654321", 2)
book3 = Book("Machine Learning", "Alice Johnson", "978-1111111111", 1)

    # 3. ÎèÑÏÑú Ï∂îÍ∞Ä
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

    # 4. ÎèÑÏÑú Î™©Î°ù Ï∂úÎ†•
library.list_books()

    # 5. ÎèÑÏÑú ÎåÄÏó¨ ÌÖåÏä§Ìä∏
library.borrow_book("978-1234567890")  # Python Programming ÎåÄÏó¨
library.borrow_book("000-0000000000")  # ÏóÜÎäî ISBN ÌÖåÏä§Ìä∏

    # 6. ÎèÑÏÑú Î∞òÎÇ© ÌÖåÏä§Ìä∏
library.return_book("978-1234567890")

    # 7. Ï±Ö Í≤ÄÏÉâ (ISBN Î∞è Ï†úÎ™©)
found_book = library.finds_book("978-0987654321")
if found_book:
        print("Found book by ISBN:", found_book)

found_book = library.finds_book("Machine Learning")
if found_book:
    print("Found book by title:", found_book)

    # 8. ÎèÑÏÑú ÏÇ≠Ï†ú ÌÖåÏä§Ìä∏
library.remove_book("978-1111111111")
library.list_books()
```

```python
class TodoItem:
    def __init__(self, title):
        self.title=title
        self.completed=False

    def mark_done(self):
        self.completed=True

    def __str__(self):
        stat="ÂÆå" if self.completed else "Êú™ÂÆå"
        return f"{self.title} [{stat}]"

```

```python
class TodoList:
    def __init__(self):
        self.CL=[]
    #TodoItemÏùÑ Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞ÄÌïúÎã§.
    def add_item(self, item):
        self.CL.append(item)
        print(f"{item.title}")

    #Î¶¨Ïä§Ìä∏ÏóêÏÑú Ìï¥Îãπ Ïù∏Îç±Ïä§Ïùò Ìï≠Î™©ÏùÑ Ï†úÍ±∞ÌïúÎã§.
    def remove_item(self, index):
        if 0<=index<len(self.CL):
            remove=self.CL.pop(index)
            print((f"ÏÇ≠Ï†úÎê®: {remove.title}"))
        else:
            print("Ïú†Ìö®ÌïòÏßÄ ÏïäÏùÄ Ïù∏Îç±Ïä§ÏûÖÎãàÎã§.")

    #Ìï¥Îãπ Ìï≠Î™©Ïùò completed ÏÉÅÌÉúÎ•º TrueÎ°ú Î≥ÄÍ≤ΩÌïúÎã§.
    def mark_done(self, index):
        if 0<=index<len(self.CL):
            self.CL[index].mark_done()
            print(f"ÏôÑÎ£åÎê®: {self.CL[index].title}")
        else:
            print("Ïú†Ìö®ÌïòÏßÄ ÏïäÏùÄ Ïù∏Îç±Ïä§ÏûÖÎãàÎã§.")

    #Î¶¨Ïä§Ìä∏Î•º ÏàúÌöåÌïòÎ©∞ Í∞Å Ìï≠Î™©Ïùò Ï†ïÎ≥¥Î•º Ï∂úÎ†•ÌïúÎã§.
    def list_items(self):
        print("\nÌòÑÏû¨ Ìï†Ïùº Î™©Î°ù:")
        if not self.CL:
            print("Ìï†Ïùº Î™©Î°ùÏù¥ ÎπÑÏñ¥ ÏûàÏäµÎãàÎã§.")
        else:
            for i,item in enumerate(self.CL, start=1):
                print(f"{i}. {item}")

            
```

```python
todo_list = TodoList()

item1 = TodoItem("Ïû•Î≥¥Í∏∞")
item2 = TodoItem("Ï≤≠ÏÜåÌïòÍ∏∞")
item3 = TodoItem("Í≥µÎ∂ÄÌïòÍ∏∞")

todo_list.add_item(item1)
todo_list.add_item(item2)
todo_list.add_item(item3)

todo_list.list_items()

    # Îëê Î≤àÏß∏ Ìï≠Î™© ÏôÑÎ£å Ï≤òÎ¶¨
todo_list.mark_done(1)
todo_list.list_items()

    # Ï≤´ Î≤àÏß∏ Ìï≠Î™© ÏÇ≠Ï†ú
todo_list.remove_item(0)
todo_list.list_items()
```
