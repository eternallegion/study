```python
class Book:
    def __init__(self, title, author, isbn, stock):
        self.title=title
        self.author=author
        self.isbn=isbn
        self.stock=stock

    def __str__(self):
        return f"{self.title} - {self.author} (ISBN: {self.isbn})"


class Library:
    def __init__(self):
        self.data={}

    def add_book(self, book):
        if book.isbn in self.data:
            EXbook=self.data[book.isbn]
            EXbook.stock+=book.stock
            print(f"Updated stock for {book.title}. New stock: {EXbook.stock}")
        else:
            self.data[book.isbn]=book
            print(f"âœ… {book.title} ì¶”ê°€ ì™„ë£Œ!")
            print(f"{book}")

    def finds_book(self, quary):
        if quary in self.data:
            return self.data[quary]
        for book in self.data.values():
            if book.title.lower()==quary.lower():
                return book
        print("Book not found.")
        return None
    

    def remove_book(self, ISBN):
        if ISBN in self.data:
            del self.data[ISBN]
            print(f"Removed book with ISBN: {ISBN}")
        else:
            return "cheak"

    def borrow_book(self, isbn):
        book=self.data.get(isbn)
        if book:
            if book.stock>0:
                book.stock-=1
                print(f"Borrowed book: {book.title}. Remaining stock: {book.stock}")
            else:
                print(f"{book.title} is not available for borrowing (out of stock).")
        else:
            print("Book not found.")

    def return_book(self,isbn):
        book=self.data.get(isbn)
        if book:
            book.stock+=1
            print(f"Returned book: {book.title}. New stock: {book.stock}")
        else:
            print("Book not found in library system.")

    def list_books(self):
        if not self.data:
            print("Library is empty.")
        else:
            print("Library Books:")
            for book in self.data.values():
                print(book)
```

```python
class TodoItem:
    def __init__(self, title):
        self.title=title
        self.completed=False

    def mark_done(self):
        self.completed=True

    def __str__(self):
        stat="å®Œ" if self.completed else "æœªå®Œ"
        return f"{self.title} [{stat}]"

```
